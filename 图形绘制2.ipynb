```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (12.8, 7.2)
## 分辨率 1500*1200
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
font = {'family': 'serif',
        'serif': 'Times New Roman',
        'weight': 'normal',
        'size': 15}
plt.rc('font', **font)

# =====================================================
x = ['GCN', 'SAGE', 'GAT', 'GraphUNet', 'GCNII']  #点的横坐标
k1 = [57.26, 49.02, 51.99, 56.07, 59.15]  #线1的纵坐标
k2 = [60.10, 59.40, 56.10, 60.20, 62.50]  #线2的纵坐标

ax1 = plt.subplot(2, 2, 1)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)

plt.title("Cora")
ax1.plot(x, k1, 's-', color='salmon', label="BaseLine")  #s-:方形
ax1.plot(x, k2, 'o-', color='skyblue', label="INC-MGS")  #o-:圆形
# plt.xlabel("Model", fontsize=15)  #横坐标名字
plt.ylabel("Accuracy(%)", fontsize=15)  #横坐标名字
plt.yticks([40, 45, 50, 55, 60, 65, 70, 75, 80, 85], fontsize=12)
plt.xticks(['GCN', 'SAGE', 'GAT', 'GraphUNet', 'GCNII'], fontsize=12)
ax1.legend(loc="best")  #图例
# =========================================================

k1 = [46.24,
      41.32,
      47.94,
      45.91,
      46.39
      ]  #线1的纵坐标
k2 = [50.70,
      46.50,
      47.40,
      49.50,
      51.50
      ]  #线2的纵坐标

ax2 = plt.subplot(2, 2, 2)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)

plt.title("Citeseer")
ax2.plot(x, k1, 's-', color='salmon', label="BaseLine")  #s-:方形
ax2.plot(x, k2, 'o-', color='skyblue', label="INC-MGS")  #o-:圆形
# plt.xlabel("Model", fontsize=15)  #横坐标名字
plt.ylabel("Accuracy(%)", fontsize=15)  #横坐标名字
plt.yticks([40, 45, 50, 55, 60, 65, 70, 75, 80, 85], fontsize=12)
plt.xticks(['GCN', 'SAGE', 'GAT', 'GraphUNet', 'GCNII'], fontsize=12)
ax2.legend(loc="best")  #图例
# ============================================================

k1 = [51.84,
      48.61,
      50.89,
      51.77,
      53.54
      ]
k2 = [53.70,
      53.40,
      52.50,
      51.30,
      51.70
      ]

ax3 = plt.subplot(2, 2, 3)
ax3.spines['top'].set_visible(False)
ax3.spines['right'].set_visible(False)

plt.title("Pubmed")
ax3.plot(x, k1, 's-', color='salmon', label="BaseLine")  #s-:方形
ax3.plot(x, k2, 'o-', color='skyblue', label="INC-MGS")  #o-:圆形
# plt.xlabel("Model", fontsize=15)  #横坐标名字
plt.ylabel("Accuracy(%)", fontsize=15)  #横坐标名字
plt.yticks([40, 45, 50, 55, 60, 65, 70, 75, 80, 85], fontsize=12)
plt.xticks(['GCN', 'SAGE', 'GAT', 'GraphUNet', 'GCNII'], fontsize=12)
ax3.legend(loc="best")  #图例
# ==============================================================

k1 = [
    76.60,
    73.81,
    79.16,
    75.21,
    81.79
]

k2 = [
    86.10,
    80.30,
    82.82,
    75.32,
    82.07
]

ax4 = plt.subplot(2, 2, 4)
ax4.spines['top'].set_visible(False)
ax4.spines['right'].set_visible(False)

plt.title("DBLP")
ax4.plot(x, k1, 's-', color='salmon', label="BaseLine")  #s-:方形
ax4.plot(x, k2, 'o-', color='skyblue', label="INC-MGS")  #o-:圆形
# plt.xlabel("Model", fontsize=15)  #横坐标名字
plt.ylabel("Accuracy(%)", fontsize=15)  #横坐标名字
plt.yticks([70, 75, 80, 85, 90, 95, 100], fontsize=12)
plt.xticks(['GCN', 'SAGE', 'GAT', 'GraphUNet', 'GCNII'], fontsize=12)
ax4.legend(loc="best")  #图例
plt.tight_layout()  # 使子图适应作图区域避免坐标轴标签信息显示混乱
# ============================= 保存
plt.savefig("引文数据.svg", dpi=300)
```
